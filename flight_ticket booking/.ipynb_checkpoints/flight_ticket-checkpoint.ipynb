```python
import pandas as pd
import tkinter as tk
from tkinter import messagebox,ttk
from pathlib import Path
from datetime import datetime
import random
```

```python
DETAILS_PATH = Path("details.xlsx")
BOOKING_PATH = Path("booking.xlsx")
TICKETS_PATH = Path("tickets.xlsx")
def load_sheet(path: Path, columns):
    if path.exists():
        try:
            df = pd.read_excel(path)
            for col in columns:
                if col not in df.columns:
                    df[col] = ""
            return df[columns]
        except Exception:
            return pd.DataFrame(columns=columns)
    return pd.DataFrame(columns=columns)
DETAILS_COLS = ["user", "mobile", "mail", "password"]
BOOKING_COLS = ["From Airport","To Airport","Flight1","Business1","Economy1","Flight2","Business2","Economy2"]
TICKET_COLS = ["Mobile","From","To","Passenger","Aadhaar","SeatType","TicketPrice","WindowPrice","MealPrice","DrinkPrice","Total",'Date',"PaymentMode",]
details = load_sheet(DETAILS_PATH, DETAILS_COLS)
booking = load_sheet(BOOKING_PATH, BOOKING_COLS)
ticket = load_sheet(TICKETS_PATH, TICKET_COLS)


root = tk.Tk()
root.title("AIR JOURNEY")
root.geometry("760x720")
root.minsize(760, 660)
root.configure(bg="#F3EFE6")
PALETTE = {"bg": "#F3EFE6","card": "#FFFDF8","muted": "#6B5E50","text": "#2B2218","accent": "#C4622D", "accent_dark": "#9D4D22", "line": "#D6C9B8"}
def setup_theme():
    style = ttk.Style(root)
    style.theme_use("clam")
    style.configure("App.TFrame",background=PALETTE["bg"])
    style.configure("Card.TFrame",background=PALETTE["card"],relief="flat")
    style.configure("Title.TLabel", font=("Georgia", 22, "bold"), foreground=PALETTE["text"], background=PALETTE["bg"])
    style.configure("Sub.TLabel",font=("Segoe UI", 10),foreground=PALETTE["muted"],background=PALETTE["bg"])
    style.configure("Field.TLabel",font=("Segoe UI", 10, "bold"),foreground=PALETTE["text"],background=PALETTE["card"])
    style.configure("Body.TLabel",font=("Segoe UI", 10),foreground=PALETTE["text"],background=PALETTE["card"])
    style.configure("App.TEntry",padding=7,fieldbackground="#FFFFFF",bordercolor=PALETTE["line"],lightcolor=PALETTE["line"],darkcolor=PALETTE["line"],font=("Segoe UI", 10))
    style.configure("App.TCombobox", padding=6,fieldbackground="#FFFFFF", bordercolor=PALETTE["line"], lightcolor=PALETTE["line"], darkcolor=PALETTE["line"], font=("Segoe UI", 10))
    style.map("Primary.TButton",background=[("active", PALETTE["accent_dark"]), ("!disabled", PALETTE["accent"])],foreground=[("!disabled", "#FFFFFF")])
    style.configure("Primary.TButton",font=("Segoe UI", 10, "bold"),padding=8,borderwidth=0)
    style.configure("Ghost.TButton",font=("Segoe UI", 10, "bold"), padding=8, foreground=PALETTE["accent_dark"],)
    style.configure("TRadiobutton",font=("Segoe UI", 10),background=PALETTE["card"],foreground=PALETTE["text"],)
    style.configure("TCheckbutton",font=("Segoe UI", 10),background=PALETTE["card"],foreground=PALETTE["text"],)
setup_theme()
main = ttk.Frame(root, style="App.TFrame")
main.pack(fill="both", expand=True, padx=24, pady=20)
root2 = ttk.Frame(root, style="App.TFrame")
main1 = ttk.Frame(root, style="App.TFrame")
main2 = ttk.Frame(root, style="App.TFrame")
main3 = ttk.Frame(root, style="App.TFrame")
main4 = ttk.Frame(root, style="App.TFrame")
main5 = ttk.Frame(root, style="App.TFrame")
data = pd.DataFrame()
current_use = ""
Passenger_names = []
Aad = []
windo = []
result = pd.DataFrame()
pl = 0
ba = ""
ml = 0
dl = 0
t = 0
total = 0
pai = "UPI"
def card(parent):
    frame = ttk.Frame(parent, style="Card.TFrame")
    frame.pack(fill="both", expand=True, padx=10, pady=10)
    frame.configure(padding=(22, 20))
    return frame
def section_header(parent, title, subtitle):
    ttk.Label(parent, text=title, style="Title.TLabel").pack(anchor="w")
    ttk.Label(parent, text=subtitle, style="Sub.TLabel").pack(anchor="w", pady=(3, 14))
def widget(frame):
    for w in frame.winfo_children():
        w.destroy()
def toggle_password():
    if pas.cget("show") == "*":
        pas.config(show="")
    else:
        pas.config(show="*")
def toggle():
    if con.cget("show") == "*":
        con.config(show="")
    else:
        con.config(show="*")
def registration():
    global name, mob, mai, pas, con
    widget(main)
    wrap = card(main)
    section_header(wrap, "Create Account", "Start booking your next air journey")
    ttk.Label(wrap, text="User Name", style="Field.TLabel").pack(fill="x", pady=(2, 3))
    name = ttk.Entry(wrap, style="App.TEntry")
    name.pack(fill="x", pady=(0, 8))
    ttk.Label(wrap, text="Mobile Number", style="Field.TLabel").pack(fill="x", pady=(2, 3))
    mob = ttk.Entry(wrap, style="App.TEntry")
    mob.pack(fill="x", pady=(0, 8))
    ttk.Label(wrap, text="Email ID", style="Field.TLabel").pack(fill="x", pady=(2, 3))
    mai = ttk.Entry(wrap, style="App.TEntry")
    mai.pack(fill="x", pady=(0, 8))
    ttk.Label(wrap, text="Create Password", style="Field.TLabel").pack(fill="x", pady=(2, 3))
    p_row = ttk.Frame(wrap, style="Card.TFrame")
    p_row.pack(fill="x", pady=(0, 8))
    pas = ttk.Entry(p_row, show="*", style="App.TEntry")
    pas.pack(side="left", fill="x", expand=True)
    ttk.Button(p_row, text="Show", style="Ghost.TButton", command=toggle_password).pack(side="left", padx=(8, 0))
    ttk.Label(wrap, text="Confirm Password", style="Field.TLabel").pack(fill="x", pady=(2, 3))
    c_row = ttk.Frame(wrap, style="Card.TFrame")
    c_row.pack(fill="x", pady=(0, 12))
    con = ttk.Entry(c_row, show="*", style="App.TEntry")
    con.pack(side="left", fill="x", expand=True)
    ttk.Button(c_row, text="Show", style="Ghost.TButton", command=toggle).pack(side="left", padx=(8, 0))
    action = ttk.Frame(wrap, style="Card.TFrame")
    action.pack(fill="x", pady=(8, 0))
    ttk.Button(action, text="Submit", style="Primary.TButton", command=sunmit).pack(side="left")
    ttk.Button(action, text="Login", style="Ghost.TButton", command=go).pack(side="left", padx=8)
def go():
    main.pack_forget()
    main1.pack(fill="both", expand=True)
    login()
def sunmit():
    global details
    user = name.get().strip()
    mobile = mob.get().strip()
    mail = mai.get().strip().lower()
    pasword = pas.get().strip()
    conform = con.get().strip()
    if mobile in list(details["mobile"].astype(str)) or mail in list(details["mail"].astype(str).str.lower()):
        messagebox.showinfo("Information", "You already have an account")
        return
    if not all([user, mobile, mail, pasword, conform]):
        messagebox.showwarning("Error", "Please fill all details")
    elif not mobile.isdigit() or len(mobile) != 10 or mobile[0] not in ["6", "7", "8", "9"]:
        messagebox.showwarning("Error", "Enter a valid mobile number")
    elif "@" not in mail or " " in mail or not mail.endswith(".com"):
        messagebox.showwarning("Error", "Enter a valid email")
    elif not any(c.isupper() for c in pasword) or not any(c in "!@#$%^&*()_+?><:}{][" for c in pasword) or not any(c.isdigit() for c in pasword):
        messagebox.showwarning("Error", "Password needs uppercase, number and special character")
    elif pasword != conform:
        messagebox.showerror("Error", "Password did not match")
    else:
        details.loc[len(details)] = [user, int(mobile), mail, pasword]
        details.to_excel(DETAILS_PATH, index=False)
        name.delete(0, tk.END)
        mob.delete(0, tk.END)
        mai.delete(0, tk.END)
        pas.delete(0, tk.END)
        con.delete(0, tk.END)
        messagebox.showinfo("Information", "Successfully registered")
def login():
    global la, ps
    widget(main1)
    wrap = card(main1)
    section_header(wrap, "Welcome Back", "Sign in with your mobile or email")
    ttk.Label(wrap, text="Mobile / Email", style="Field.TLabel").pack(fill="x", pady=(2, 3))
    la = ttk.Entry(wrap, style="App.TEntry")
    la.pack(fill="x", pady=(0, 8))
    ttk.Label(wrap, text="Password", style="Field.TLabel").pack(fill="x", pady=(2, 3))
    p_row = ttk.Frame(wrap, style="Card.TFrame")
    p_row.pack(fill="x", pady=(0, 12))
    ps = ttk.Entry(p_row, show="*", style="App.TEntry")
    ps.pack(side="left", fill="x", expand=True)
    def toggl():
        if ps.cget("show") == "*":
            ps.config(show="")
        else:
            ps.config(show="*")
    ttk.Button(p_row, text="Show", style="Ghost.TButton", command=toggl).pack(side="left", padx=(8, 0))
    action = ttk.Frame(wrap, style="Card.TFrame")
    action.pack(fill="x", pady=(8, 0))
    ttk.Button(action, text="Login", style="Primary.TButton", command=golog).pack(side="left")
    ttk.Button(action, text="Sign Up", style="Ghost.TButton", command=back_to_signup).pack(side="left", padx=8)
def back_to_signup():
    main1.pack_forget()
    main.pack(fill="both", expand=True)
    registration()
def golog():
    global data
    use = la.get().strip()
    pa = ps.get().strip()
    if not use:
        messagebox.showwarning("Error", "Please enter login details")
        return
    if use not in list(details["mail"].astype(str)) and use not in list(details["mobile"].astype(str)):
        messagebox.showwarning("Error", "Sign up your account first")
        la.delete(0, tk.END)
        ps.delete(0, tk.END)
        return
    data = details.loc[(details["mail"].astype(str) == use) | (details["mobile"].astype(str) == use)]
    if not data.empty and pa in data["password"].values:
        la.delete(0, tk.END)
        ps.delete(0, tk.END)
        main1.pack_forget()
        main2.pack(fill="both", expand=True)
        log()
        root2.place(relx=1.0, x=-20, y=14, anchor="ne")
    else:
        messagebox.showwarning("Error", "Incorrect password")
        ps.delete(0, tk.END)
def log():
    global from1, to, current_use
    widget(root2)
    current_use = data["user"].values[0]
    badge = ttk.Frame(root2, style="Card.TFrame", padding=(10, 6))
    badge.pack()
    ttk.Label(badge, text=f"User: {current_use}", style="Body.TLabel").pack()
    widget(main2)
    wrap = card(main2)
    section_header(wrap, "Book Flight", "Choose your route and continue")
    ttk.Label(wrap, text="From", style="Field.TLabel").pack(fill="x", pady=(2, 3))
    airport_list = sorted(list(set(booking["From Airport"].tolist() + booking["To Airport"].tolist())))
    from1 = ttk.Combobox(wrap,values=airport_list,state="readonly",style="App.TCombobox")
    from1.pack(fill="x", pady=(0, 8))
    ttk.Label(wrap, text="To", style="Field.TLabel").pack(fill="x", pady=(2, 3))
    to = ttk.Combobox(wrap,values=airport_list,state="readonly",style="App.TCombobox")
    to.pack(fill="x", pady=(0, 12))
    actions = ttk.Frame(wrap, style="Card.TFrame")
    actions.pack(fill="x")
    ttk.Button(actions, text="Show Flights", style="Primary.TButton", command=book).pack(side="left")
    ttk.Button(actions, text="My Tickets", style="Ghost.TButton", command=show_df).pack(side="left", padx=8)
    ttk.Button(actions, text="Logout", style="Ghost.TButton", command=logout).pack(side="right")
def logout():
    root2.place_forget()
    main2.pack_forget()
    main1.pack(fill="both", expand=True)
    login()
def show_df():
    main2.pack_forget()
    main5.pack(fill="both", expand=True)
    widget(main5)
    wrap = card(main5)
    section_header(wrap, "My Tickets", "All your booked tickets")
    current_user = int(data["mobile"].values[0])
    user_df = ticket[ticket["Mobile"].astype(int) == current_user]
    if user_df.empty:
        ttk.Label(wrap, text="No tickets found", style="Body.TLabel").pack(anchor="w")
    else:
        cols = list(user_df.columns)
        tree = ttk.Treeview(wrap, columns=cols, show="headings", height=10)
        for c in cols:
            tree.heading(c, text=c)
            tree.column(c, width=96, anchor="center")
        for _, row in user_df.iterrows():
            tree.insert("", "end", values=list(row))
        tree.pack(fill="both", expand=True, pady=(4, 8))
        scroll = ttk.Scrollbar(wrap, orient="vertical", command=tree.yview)
        tree.configure(yscroll=scroll.set)
        scroll.place(relx=1.0, rely=0.52, anchor="e", relheight=0.6)
    ttk.Button(wrap, text="Back", style="Ghost.TButton", command=lambda: [main5.pack_forget(), main2.pack(fill="both", expand=True), log()]).pack(anchor="w")
def book():
    global result, p
    from2 = from1.get().strip()
    to2 = to.get().strip()
    if not from2 or not to2:
        messagebox.showwarning("Error", "Please select route")
        return
        if from2 == to2:
            messagebox.showinfo("Place Error", "Source and destination cannot be same")
            return
    result = booking[(booking["From Airport"] == from2) & (booking["To Airport"] == to2)]
    if result.empty:
        messagebox.showwarning("Error", "Route not available")
        return
    main2.pack_forget()
    main3.pack(fill="both", expand=True)
    widget(main3)
    wrap = card(main3)
    section_header(wrap, "Available Flights", f"{from2} to {to2}")
    flights = list(result["Flight1"]) + list(result["Flight2"])
    flights = [f for f in flights if pd.notna(f)]
    ttk.Label(wrap, text="Choose Flight", style="Field.TLabel").pack(fill="x", pady=(2, 3))
    br = ttk.Combobox(wrap, values=flights, state="readonly", style="App.TCombobox")
    br.pack(fill="x", pady=(0, 8))
    ttk.Label(wrap, text="Passengers", style="Field.TLabel").pack(fill="x", pady=(2, 3))
    p = ttk.Combobox(wrap, values=[1, 2, 3], state="readonly", style="App.TCombobox")
    p.pack(fill="x", pady=(0, 12))
    def close():
        global ba, pl
        ba = br.get().strip()
        val = p.get().strip()
        if not ba or not val:
            messagebox.showwarning("Error", "Please fill all details")
            return
        pl = int(val)
        show_seat()
    actions = ttk.Frame(wrap, style="Card.TFrame")
    actions.pack(fill="x")
    ttk.Button(actions, text="Next", style="Primary.TButton", command=close).pack(side="left")
    ttk.Button(actions, text="Back", style="Ghost.TButton", command=lambda: [main3.pack_forget(), main2.pack(fill="both", expand=True), log()]).pack(side="left", padx=8)
def show_seat():
    global Passenger_names, Aad, windo, meal, drink, se
    Passenger_names = []
    Aad = []
    windo = []
    main3.pack_forget()
    main4.pack(fill="both", expand=True)
    widget(main4)
    wrap = card(main4)
    section_header(wrap, "Passenger Details", "Select seat and add all passengers")
    ttk.Label(wrap, text="Ticket Type", style="Field.TLabel").pack(anchor="w")
    se = tk.StringVar(value="Economy")
    ttk.Radiobutton(wrap, text="Economy", variable=se, value="Economy").pack(anchor="w")
    ttk.Radiobutton(wrap, text="Business", variable=se, value="Business").pack(anchor="w", pady=(0, 8))
    for i in range(pl):
        band = ttk.Frame(wrap, style="Card.TFrame", padding=(10, 8))
        band.pack(fill="x", pady=(5, 5))
        ttk.Label(band, text=f"Passenger {i + 1} Name", style="Field.TLabel").pack(fill="x", pady=(0, 3))
        book_n = ttk.Entry(band, style="App.TEntry")
        book_n.pack(fill="x", pady=(0, 6))
        Passenger_names.append(book_n)
        ttk.Label(band, text=f"Passenger {i + 1} Aadhaar", style="Field.TLabel").pack(fill="x", pady=(0, 3))
        book_a = ttk.Entry(band, style="App.TEntry")
        book_a.pack(fill="x", pady=(0, 6))
        Aad.append(book_a)
        agree = tk.IntVar(value=0)
        ttk.Checkbutton(band, text="Window seat", variable=agree).pack(anchor="w")
        windo.append(agree)
    ttk.Label(wrap, text="Dining", style="Field.TLabel").pack(anchor="w", pady=(8, 2))
    meal = tk.IntVar(value=0)
    drink = tk.IntVar(value=0)
    ttk.Checkbutton(wrap, text="Food", variable=meal).pack(anchor="w")
    ttk.Checkbutton(wrap, text="Drinks", variable=drink).pack(anchor="w")
    action = ttk.Frame(wrap, style="Card.TFrame")
    action.pack(fill="x", pady=(10, 0))
    ttk.Button(action, text="Confirm", style="Primary.TButton", command=conform).pack(side="left")
    ttk.Button(action, text="Back", style="Ghost.TButton", command=lambda: [main4.pack_forget(), main3.pack(fill="both", expand=True)]).pack(side="left", padx=8)
def conform():
    global Passanger, Aadar, window
    Passanger = [e.get().strip() for e in Passenger_names]
    Aadar = [e.get().strip() for e in Aad]
    window = [e.get() for e in windo]
    if "" in Passanger or "" in Aadar:
        messagebox.showwarning("Error", "Please enter passenger details")
        return
    if len(set(Aadar)) < len(Aadar):
        messagebox.showwarning("Error", "Aadhaar numbers cannot be same")
        return
    if any((not a.isdigit()) or len(a) != 12 for a in Aadar):
        messagebox.showwarning("Error", "Enter valid 12-digit Aadhaar number")
        return
    main4.pack_forget()
    main5.pack(fill="both", expand=True)
    payment()
def payment():
    global cash, pay, total, t, ml, dl
    widget(main5)
    wrap = card(main5)
    section_header(wrap, "Payment Session", "Review fare before confirmation")
    seat = se.get()
    count = window.count(1)
    total = 0
    ml = 0
    dl = 0
    t = 0
    if seat == "Economy":
        if ba in result["Flight1"].values:
            t = int(result["Economy1"].values[0])
        elif ba in result["Flight2"].values:
            t = int(result["Economy2"].values[0])
    elif seat == "Business":
        if ba in result["Flight1"].values:
            t = int(result["Business1"].values[0])
        elif ba in result["Flight2"].values:
            t = int(result["Business2"].values[0])
    total += t * pl
    if meal.get() == 1:
        ml = pl
        total += 25 * pl
    if drink.get() == 1:
        dl = pl
        total += 15 * pl
    total += count * 7
    ttk.Label(wrap, text=f"Flight: {ba}", style="Body.TLabel").pack(anchor="w")
    ttk.Label(wrap,text=f"{result['From Airport'].values[0]} -> {result['To Airport'].values[0]}",style="Body.TLabel").pack(anchor="w")
    ttk.Label(wrap, text=f"Seat Type: {seat}", style="Body.TLabel").pack(anchor="w", pady=(0, 6))
    ttk.Label(wrap, text=f"Ticket: {t} x {pl} = ${t * pl}", style="Body.TLabel").pack(anchor="w")
    ttk.Label(wrap, text=f"Window: 7 x {count} = ${7 * count}", style="Body.TLabel").pack(anchor="w")
    ttk.Label(wrap, text=f"Meal: 25 x {ml} = ${25 * ml}", style="Body.TLabel").pack(anchor="w")
    ttk.Label(wrap, text=f"Drinks: 15 x {dl} = ${15 * dl}", style="Body.TLabel").pack(anchor="w", pady=(0, 8))
    ttk.Label(wrap, text=f"Total: ${total}", style="Field.TLabel").pack(anchor="w", pady=(0, 10))
    pay = tk.StringVar(value="UPI")
    ttk.Radiobutton(wrap, text="UPI", variable=pay, value="UPI").pack(anchor="w")
    ttk.Radiobutton(wrap, text="Debit/Credit Card", variable=pay, value="Card").pack(anchor="w")
    cash = ttk.Entry(wrap, style="App.TEntry")
    cash.pack(fill="x", pady=(8, 10))
    def finish():
        global pai
        fi = cash.get().strip()
        pai = pay.get()
        if not fi:
            messagebox.showwarning("Error", "Fill payment details")
            return
        if pai == "UPI":
            messagebox.showinfo("Success", f"Ticket confirmed for {', '.join(Passanger)}\nPaid {total} using {pai}: {fi}")
        else:
            messagebox.showinfo("Success", f"Ticket confirmed for {', '.join(Passanger)}\nPaid {total} using {pai}: ****{fi[-4:]}")
        show_ticket()
    action = ttk.Frame(wrap, style="Card.TFrame")
    action.pack(fill="x")
    ttk.Button(action, text="Confirm", style="Primary.TButton", command=finish).pack(side="left")
    ttk.Button(action, text="Cancel", style="Ghost.TButton", command=lambda: [main5.pack_forget(), main4.pack(fill="both", expand=True)]).pack(side="left", padx=8)
def show_ticket():
    global ticket
    seat = se.get()
    temp_mobile = [int(data["mobile"].values[0])] * pl
    temp_from = [result["From Airport"].values[0]] * pl
    temp_to = [result["To Airport"].values[0]] * pl
    temp_type = [seat] * pl
    temp_meal = [(ml * 25) / pl] * pl
    temp_drinks = [(dl * 15) / pl] * pl
    temp_pay = [pai] * pl
    temp_t = [t] * pl
    issued_on = datetime.now().strftime("%d-%b-%Y %I:%M %p")
    date=[issued_on]*pl
    for i in range(pl):
        ticket.loc[len(ticket)] = [temp_mobile[i],temp_from[i],temp_to[i],Passanger[i],Aadar[i],temp_type[i],
                temp_t[i],window[i] * 7,temp_meal[i],temp_drinks[i],temp_t[i] + (window[i] * 7) + temp_meal[i] + temp_drinks[i],date[i],temp_pay[i],]
    ticket.to_excel(TICKETS_PATH, index=False)
    widget(main5)
    wrap = card(main5)
    section_header(wrap, "Air Ticket", "Booking confirmed")
    section_header(wrap, "E-Ticket / Boarding Pass", "Booking confirmed")
    header = ttk.Frame(wrap, style="Card.TFrame", padding=(12, 10))
    header.pack(fill="x", pady=(0, 8))
    pnr = f"AJ{random.randint(100000, 999999)}"
    ttk.Label(header, text="AIR JOURNEY", style="Field.TLabel").pack(anchor="w")
    ttk.Label(header,text=f"PNR: {pnr}    Ticket No: {random.randint(1000000000, 9999999999)}",style="Body.TLabel").pack(anchor="w")
    ttk.Label(header, text=f"Issued: {issued_on}", style="Body.TLabel").pack(anchor="w")
    route = ttk.Frame(wrap, style="Card.TFrame", padding=(12, 10))
    route.pack(fill="x", pady=(0, 8))
    ttk.Label(route,text=f"{result['From Airport'].values[0]}  ->  {result['To Airport'].values[0]}",style="Field.TLabel").pack(anchor="w")
    ttk.Label(route, text=f"Flight: {ba}", style="Body.TLabel").pack(anchor="w")
    ttk.Label(route, text=f"Class: {seat}", style="Body.TLabel").pack(anchor="w")
    ttk.Label(route, text=f"Passengers: {len(Passanger)}", style="Body.TLabel").pack(anchor="w")
    pax = ttk.Frame(wrap, style="Card.TFrame", padding=(12, 10))
    pax.pack(fill="x", pady=(0, 8))
    ttk.Label(pax, text="Passenger Details", style="Field.TLabel").pack(anchor="w")
    for i, nm in enumerate(Passanger, start=1):
        seat_pref = "Window" if window[i - 1] else "Regular"
        ttk.Label(pax,text=f"{i}. {nm} | Aadhaar:{'X'*8} {Aadar[i - 1][-4:]} | Seat Pref: {seat_pref}",style="Body.TLabel",).pack(anchor="w")
    fare = ttk.Frame(wrap, style="Card.TFrame", padding=(12, 10))
    fare.pack(fill="x", pady=(0, 10))
    ttk.Label(fare, text="Fare Summary", style="Field.TLabel").pack(anchor="w")
    ttk.Label(fare, text=f"Base Fare ({pl} x ${t}): ${pl * t}", style="Body.TLabel").pack(anchor="w")
    ttk.Label(fare, text=f"Window Seats: ${window.count(1) * 7}", style="Body.TLabel").pack(anchor="w")
    ttk.Label(fare, text=f"Meals: ${ml * 25}", style="Body.TLabel").pack(anchor="w")
    ttk.Label(fare, text=f"Drinks: ${dl * 15}", style="Body.TLabel").pack(anchor="w")
    ttk.Label(fare, text=f"Payment Mode: {pay.get()}", style="Body.TLabel").pack(anchor="w")
    ttk.Label(fare, text=f"Total Paid: ${total}", style="Field.TLabel").pack(anchor="w", pady=(4, 0))
    ttk.Button(wrap, text="Book Another Ticket", style="Primary.TButton", command=reset_all).pack(anchor="w")
def reset_all():
    global Passenger_names, Aad, windo
    for frame in [main2, main3, main4, main5]:
        widget(frame)
    Passenger_names = []
    Aad = []
    windo = []
    main3.pack_forget()
    main4.pack_forget()
    main5.pack_forget()
    main2.pack(fill="both", expand=True)
    log()
registration()
root.mainloop()
```
